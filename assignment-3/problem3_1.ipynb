```python
import nbdime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
protein_data = pd.read_csv("assignment3-data.csv", index_col = False)
angle_data = protein_data[['phi', 'psi']]

number_of_colors = len(angle_data)
colors = np.random.rand(number_of_colors)
# plots very small points due to the large number of data points
plt.scatter(angle_data['phi'], angle_data['psi'], c = colors, s=0.02)
plt.xlabel("phi")
plt.ylabel("psi")
plt.title("Figure 1: Angles of rotation for a protein chain")
plt.show()

heatmap_of_angles, phi_edge, psi_edge = np.histogram2d(angle_data["phi"], angle_data["psi"], bins=30)
our_extent = [phi_edge[0], phi_edge[-1], psi_edge[0], psi_edge[-1]]

plt.imshow(heatmap_of_angles.T, extent=our_extent, cmap='magma', origin='lower')
plt.show()
```
