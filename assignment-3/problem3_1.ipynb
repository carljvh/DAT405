```python
import nbdime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
protein_data = pd.read_csv("assignment3-data.csv", index_col = False)
angle_data = protein_data[['phi', 'psi']]

N = len(angle_data)
colors = np.random.rand(N)
plt.scatter(angle_data['phi'], angle_data['psi'], c = colors, s=1)
plt.xlabel("phi")
plt.ylabel("psi")
plt.title("Figure 1: Angles of rotation for a protein chain")
plt.show()

heatmap, xedges, yedges = np.histogram2d(angle_data["phi"], angle_data["psi"], bins=50)
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

plt.clf()
plt.imshow(heatmap.T, extent=extent, origin='lower')
plt.show()
```
