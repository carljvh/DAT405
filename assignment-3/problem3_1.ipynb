```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statistics as st
from sklearn import datasets
import os
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

protein_data = pd.read_csv("assignment3-data.csv", index_col = False)

phi_data = protein_data["phi"]
psi_data = protein_data["psi"]

N = len(phi_data)
colors = np.random.rand(N)
plt.scatter(phi_data, psi_data, c = colors)
plt.xlabel("phi")
plt.ylabel("psi")
plt.title("Figure 1: Angels of rotation for a protein chain")
plt.show()

heatmap, xedges, yedges = np.histogram2d(phi_data, psi_data, bins=50)
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

plt.clf()
plt.imshow(heatmap.T, extent=extent, origin='lower')
plt.show()
```
